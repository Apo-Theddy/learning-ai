```python
# Trabajando con el CSV Nombres

df_base.info()
```

```python
df_mary = df_base[
    (df_base["sex"] == "M") & (df_base["name"] == "Mary")
]

# Consultamos los primeros cinco registros
df_mary.head()
```

```python
# Generar un nuevo df
nuevo_df = df_mary[["state","quantity"]]
nuevo_df.head()
```

```python
# Cantidad de hombres Mary han habido, podemos generar la sumatoria de forma similar a la estadistica descriptiva
nuevo_df[["quantity"]].sum()
```

```python
# Agrupar por estado utilizando la funcion sum
estados_df = nuevo_df.groupby("state").agg("sum")
estados_df.head()
```

```python
# Consultar todos los registros de nuevo DataFrame
estados_df
```

```python
# Reto
# Obtener la cantidad promedio de Mary por estado usando estados_df
estados_df = nuevo_df.groupby("state").agg("mean")
estados_df
```

```python
# Ordenar el set de datos por la mayor cantidad

estados_df.sort_values("quantity",ascending=False)
```

```python
# Por ello es posible acceder por estados
estados_df.loc["CA"]
```

```python
# Para convertir el indice en una columna del df, se puede generar uno nuevo aleatorio con reset
# Nota al uso de inplace=True para modificar el df
estados_df.reset_index(inplace=True)
estados_df
```

```python
# Exportar a un csv
estados_df.to_excel("estados_mary.xlsx",index=None)
```
