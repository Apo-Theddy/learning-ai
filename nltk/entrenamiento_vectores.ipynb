```python
from gensim.models import Word2Vec
import string
from PyPDF2 import PdfReader
```

```python
# carga del documento
pdf_reader = PdfReader("codigo-limpio.pdf")
documento = ""
for page_num in pdf_reader.pages:
 documento += page_num.extract_text()
```

```python
len(documento)
```

```python
## Preprocesamiento de datos
oraciones = documento.split(".")
len(oraciones)
```

```python
oraciones_limpias = []
for oracion in oraciones:
 # elimnar puntuacion y dividir por espacios
 tokens = oracion.translate(str.maketrans("","",string.punctuation)).split()
 tokens = [word.lower() for word in tokens if word.isalpha()]
 if tokens:
  oraciones_limpias.append(tokens)
 
```

```python
# Entrenamiento del modelo Word2Vec

model = Word2Vec(sentences=oraciones_limpias,vector_size=500,window=5,min_count=1,workers=8)
```

```python
vector = model.wv["limpio"]
vector
```

```python
palabras_cercanas = model.wv.most_similar("limpio",topn=10)
palabras_cercanas
```

```python
# guardar modelo
model.save("codigo-limpio.model")

# cargar el modelo
modelo_cargado = Word2Vec.load("codigo-limpio.model")

model.wv.save_word2vec_format("codigo-limpio-embedding.txt",binary=False)
model.wv.save_word2vec_format("codigo-limpio-embedding.bin",binary=True)
```

```python
from gensim.models import KeyedVectors

embedding_cargados = KeyedVectors.load_word2vec_format("codigo-limpio-embedding.bin",binary=True)

embedding_cargados
```

```python
def analogias(v1,v2,v3):
 similitud = embedding_cargados.most_similar(positive=[v1,v3],negative=[v2])
 print(f"{v1} es a {v2} como {similitud[0][0]} es a {v3}")
```

```python
analogias("limpio","java","nada")
```
